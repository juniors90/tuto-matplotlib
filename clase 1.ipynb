```python
import matplotlib.pyplot as plt
import numpy as np
```

definimos las varia

```python
x = np.linspace(0, 2 * np.pi, 200)
y = np.cos(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
plt.grid()
fig.savefig("cos.png")
plt.show()
```

```python
# Data for plotting
t = np.arange(0.0, 5.0, 0.025)
s = 1 + np.cos(2 * np.pi * t)

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='tiempo (s)', ylabel='voltaje (mV)',
       title='Turorial')
ax.grid()
plt.show()
```
